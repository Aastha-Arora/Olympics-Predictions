```python
import pandas as pd
```

# Data Pre-Processing

```python
# Load training data
train_data = pd.read_csv("train.csv")
train_data 
```

```python
# Load testing data
test_data = pd.read_csv("test.csv")

test_data
```

```python
train_data = pd.DataFrame(train_data)
train_data.describe().columns
```

```python
# Assign X and y for test and training data
y_train = train_data["Total_Medals"]
X_train_features = train_data[["Year", "Athletes", "Event","Sport"]]
X_train = X_train_features.values


X_test_features = test_data[["Year", "Athletes", "Event","Sport"]]
X_test = X_test_features.values
```

# Create and Train(Fit) Model

```python
 # Create a LinearRegression model and fit it to the scaled training data

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
```
