```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Summer_Olympics_Host.csv")
df.head(5)
```

```python
df_usa = df.loc[df['Country']=='USA'].reset_index(drop=True)
df_usa
```

```python
# new_row = pd.Series(data={'Year':2020, 'Country':'USA', 'Athletes':340, 
#                           'Host':0, 'Sport':25, 'Event':191}, name='0')
# new_row
```

```python
# df_usa = df_usa.append(new_row, ignore_index=True)
# df_usa
```

```python
df_usa['Athletes per Event'] = df_usa['Athletes']/df_usa['Event']
```

```python
df_usa
```

```python
usa_train = df_usa.loc[df_usa['Year']<=2000]
usa_test = df_usa.loc[(df_usa['Year']>2000)]
# usa_test = df_usa.loc[(df_usa['Year']==2012) | (df_usa['Year']==2016)]
```

```python
usa_train
```

```python
X_train = usa_train[['Year','Athletes per Event','Host']]
X_train
```

```python
y_train = usa_train["Total_Medals"]
y_train = y_train.values.reshape(-1,1)
y_train
```

```python
print(X_train.shape, y_train.shape)
```

```python
usa_test
```

```python
# new_row = pd.Series(data={'Year':2020, 'Country':'USA', 'Athletes':340, 
#                           'Host':0, 'Sport':25, 'Event':191}, name='0')
# new_row
```

```python
# usa_test = usa_test.append(new_row, ignore_index=False)
# usa_test
```

```python
X_test = usa_test[['Year','Athletes per Event', 'Host']]
X_test
```

```python
# X_test_2020 = X_test.loc[28]
# X_test_2020
```

```python
# X_test = X_test.drop([28])
# X_test
```

```python
y_test = usa_test["Total_Medals"]
# y_test = y_test.values.reshape(-1,1)
# y_test = y_test.drop([28])
y_test = y_test.values.reshape(-1,1)
y_test
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
# print('Weight coefficients: ', model.coef_)
# print('y-axis intercept: ', model.intercept_) 
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to predict a value
predicted = model.predict(X_test)
predicted
# Score the prediction with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Residuals for the Training and Testing data
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data");
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_train.min(), xmax=y_train.max())
plt.title("Residual Plot")
```

```python
import numpy as np
pd.DataFrame({"Predicted": (np.ravel(predicted)), "Actual": (np.ravel(y_test)), "Error": np.ravel(predicted) - np.ravel(y_test)})
```

```python
# LASSO model
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=.01).fit(X_train, y_train)

predictions = lasso.predict(X_test)

MSE = mean_squared_error(y_test, predictions)
r2 = lasso.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
# Ridge model
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=.01).fit(X_train, y_train)

predictions = ridge.predict(X_test)

MSE = mean_squared_error(y_test, predictions)
r2 = ridge.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
# ElasticNet model
from sklearn.linear_model import ElasticNet

elasticnet = ElasticNet(alpha=.01).fit(X_train, y_train)

predictions = elasticnet.predict(X_test)

MSE = mean_squared_error(y_test, predictions)
r2 = elasticnet.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
```
