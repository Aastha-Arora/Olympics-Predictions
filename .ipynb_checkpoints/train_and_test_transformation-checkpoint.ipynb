```python
import pandas as pd
```

```python
# Load Data - Summer Olympics without HOST

df = pd.read_csv('Resources/Summer_Olympics_Host.csv')

df.head(3)
```

```python
# Filter for country: USA

country_df = df[df['Country'] == 'USA']
country_df.head(3)
```

# Test data

```python
#  Ceate test data
test_df = country_df[country_df['Year'] > 2012]
test_df = pd.DataFrame(test_df)

# Dropping Total Medal count to prep for predictions
# test_df = test_df.drop(["Total_Medals","Gold","Silver","Bronze"], axis=1)
test_df = test_df.drop(["Total_Medals"], axis=1)

test_df
```

```python
# Adding 2020 to prep for future prediction after hyperparameter tuning and training

new_row = pd.Series(data={'Year':2020, 'Country':'USA', 'Athletes':0, 'Event': 0, 'Sport':0, 'Host':0}, name='0')
test_df = test_df.append(new_row, ignore_index=False)

test_df
```

```python
#  Save into a csv file
test_df.to_csv('Resources/test.csv', index=False)
```

# Train Data

```python
#  Create train data
train_df = country_df[country_df['Year'] <= 2012]
train_df = pd.DataFrame(train_df)
train_df
```

```python
#  Save into a csv file
train_df.to_csv('Resources/train.csv', index=False)
```
