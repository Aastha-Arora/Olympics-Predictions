```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
athlete_events = pd.read_csv('../CSV for ML models/athlete_events.csv')
```

```python
athlete_events.head()
```

```python
filter_data = athlete_events[["Sex", "Age", "Height", "Weight", "Team", "Year", "Season", "Sport", "Event", "Medal"]]
```

```python
winter_data = filter_data[filter_data["Season"] == "Winter"]
summer_data = filter_data[(filter_data["Season"] == "Summer")]
```

```python
summer_sports = summer_data["Sport"].unique()
```

```python
summer_sports = summer_data["Sport"].unique()
```

```python
for sport in summer_sports:
    try:
        Gymnastics_MF = summer_data[(summer_data["Sport"] == f'{sport}') & (summer_data["Sex"] == "M")]
        Gymnastics_MF = Gymnastics_MF[["Age", "Height", "Weight", "Medal"]]
        Gymnastics_MF = Gymnastics_MF.dropna(subset=['Height', 'Weight']).reset_index(drop = True)
        Gymnastics_MF = Gymnastics_MF.replace("Gold", 1)
        Gymnastics_MF = Gymnastics_MF.replace("Silver", 1)
        Gymnastics_MF = Gymnastics_MF.replace("Bronze", 1)
        Gymnastics_MF["Medal"] = Gymnastics_MF["Medal"].fillna(0)
        Gymnastics_MF = Gymnastics_MF.dropna()
        Gymnastics_MF["Medal"].unique()

        print(sport)
        
        
        target = Gymnastics_MF["Medal"]
        target_names = ["Medals", "None"]
        data = Gymnastics_MF.drop("Medal", axis=1)
        feature_names = data.columns
        data.head()

        X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
        clf = tree.DecisionTreeClassifier()
        clf = clf.fit(X_train, y_train)
        clf.score(X_test, y_test)

        rf = RandomForestClassifier(n_estimators=200)
        rf = rf.fit(X_train, y_train)
        print("Testing Data Score:", rf.score(X_test, y_test))
        sorted(zip(rf.feature_importances_, feature_names), reverse=True)
        print("----------------------------------------------------")
    except:
        print("An exception occurred")
        print("----------------------------------------------------")
```
