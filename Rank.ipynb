```python
import pandas as pd
```

```python
olympics_data = pd.read_csv('olympics_data.csv')
print(olympics_data.shape)
olympics_data.head()
```

## Summer Olympics
---

```python
summer_df = olympics_data[olympics_data['Season'] == 'Summer']
print(summer_df.shape)
summer_df.head()
```

```python
print(len(summer_df['Year'].unique()))
print(sorted(summer_df['Year'].unique()))
```

```python
summer_df = summer_df[['Country', 'Year', 'Medal']]
print(summer_df.shape)
summer_df.head()
```

```python
summer_df = summer_df.dropna()
print(summer_df.shape)
summer_df.head()
```

```python
summer_df = summer_df.groupby(['Year','Country'], sort = False).count().reset_index()
print(summer_df.shape)
summer_df.head()
```

```python
summer_df = summer_df.sort_values(['Year', 'Medal', 'Country'], ascending = [True, False, True])
print(summer_df.shape)
summer_df.head(10)
```

```python
summer_df["Rank"] = summer_df.groupby("Year")["Medal"].rank("dense", ascending=False)
print(summer_df.shape)
summer_df.head(10)
```

```python
summer_df = summer_df[summer_df['Rank'] == 1]
print(summer_df.shape)
summer_df.head()
```

```python
summer_rank_df = summer_df.groupby(['Country']).count()['Rank'].reset_index().sort_values(['Rank'], ascending = False)
print(summer_rank_df.shape)
summer_rank_df
```
