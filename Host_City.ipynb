```python
import pandas as pd
```

```python
df= pd.read_csv("olympics_data.csv")
df.head(2)
```

```python
df_summer = df.loc[df['Season']=="Summer"]
```

```python
host_df = df_summer.groupby(['Year','City']).nunique()[["Country","Sport","Event"]].reset_index()
host_df.head(2)
```

```python
host_df.to_csv('host.csv', index=False)
```

### Country names were added manually using data from Wikipedia
https://en.wikipedia.org/wiki/List_of_Olympic_Games_host_cities

```python
host_countries = pd.read_csv("host_country.csv")
```

```python
host_countries.head(2)
```

```python
host_countries = host_countries[["Year", "Host_Country"]]
host_countries.head(2)
```

### Adding host country to summer_olympics data

```python
summer_olympics = pd.read_csv("Summer_Olympics.csv")
```

```python
summer_olympics.head(2)
```

```python
summer_olympics_host = pd.merge(summer_olympics, host_countries, on="Year", how="outer")
summer_olympics_host.head(5)
```

```python
summer_olympics_host['is_host'] = summer_olympics_host['Country'] == summer_olympics_host['Host_Country']
```

```python
summer_olympics_host['Host'] = [1 if x is True else 0 for x in summer_olympics_host['is_host']] 
```

```python
summer_olympics_host.head(5)
```

```python
summer_olympics_host = summer_olympics_host.drop(['is_host', 'Host_Country'], axis=1)
```

```python
summer_olympics_host.to_csv('Summer_Olympics_Host.csv', index = False)
```
